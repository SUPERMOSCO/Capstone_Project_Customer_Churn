**Q.1. What are the airlines in the dataset, accompanied by their frequencies?**

```python
data['airline'].nunique()
```

```python
data['airline'].unique()
```

```python
data['airline'].value_counts(ascending=True)
```

```python
# OR
data.airline.value_counts(ascending=True)
```

```python
# Showing all the Airlines with their Number of Flights in Horizontal Bar Graph
plt.figure(figsize=(10,6))
data['airline'].value_counts(ascending=True).plot.barh(color=['green','lightblue'])
plt.title('Airline with Frequency')
plt.xlabel('Frequency')
plt.ylabel('Airline')
plt.show()
```

**Q.2. Show Bar Graphs representing the Departure Time & Arrival Time**

```python
data.head()
```

```python
data['departure_time'].value_counts(ascending=True)  # From the above result, it can be deduced that most flights depart in the morning and least number of flights depart at late night.
```

```python
data['arrival_time'].value_counts(ascending=True)  # From the above result, it can be deduced that most flights arrived at night and least number of flights arrived at late night.
```

```python
# Viewing our result in a graph
plt.figure(figsize=(16,4))
plt.subplot(1,2,1)
plt.bar(data['departure_time'].value_counts().index,data['departure_time'].value_counts().values,color=['red','orange'])
plt.title('Departure Time of Flight')
plt.xlabel('Departure Time')
plt.ylabel('Frequency')


plt.subplot(1,2,2)
# Viewing our result in a graph
plt.bar(data['arrival_time'].value_counts().index, data['arrival_time'].value_counts().values,color=['blue','purple'])
plt.title('Arrival Time of Flight')
plt.xlabel('Arrival Time')
plt.ylabel('Frequency')
plt.show()
```

**Q.3. Show Bar Graphs representing the Source City & Destination City**

```python
data.head()
```

```python
data['source_city'].value_counts(ascending=True)  # It can be said from the data below that most of Airines are leaving from Delhi and the least Airlines are leaving from Chennai
```

```python
data['destination_city'].value_counts(ascending=True) # It can be deduced that Mumbai is the destination of most of the flight and Chennai is the least destination of the Airline.
```

```python
plt.figure(figsize=(16,4))

plt.subplot(1,2,1)

plt.barh(data['source_city'].value_counts().index, data['source_city'].value_counts().values, color=['lightgreen','lightblue'])
plt.title('Source City of Airline')
plt.xlabel('No of Flight')
plt.ylabel('Source City')

plt.subplot(1,2,2)

plt.barh(data['destination_city'].value_counts().index, data['destination_city'].value_counts().values, color=['orange','lightblue'])
plt.title('Destination City of Airline')
plt.xlabel('No of Flight')
plt.ylabel('Destination City')
```

**Q.4. Does price varies with airlines ?**

```python
data.head()
```

```python
data.groupby('airline')['price'].mean()
```

```python
# showing it in a bar chart

plt.figure(figsize=(12,5))
sns.catplot(x='airline', y='price',data=data, kind='bar', palette='rocket', hue='class')
plt.title('Price of Airline')
plt.xlabel('Airline')
plt.show()
```

**Q.5. Does ticket price change based on the departure time and arrival time?**

```python
data.head()
```

```python
data.groupby('departure_time')['price'].mean()
```

```python
data.groupby('arrival_time')['price'].mean()
```

```python
plt.figure(figsize=(12,5))
sns.catplot(x='departure_time',y='price',data=data, kind='bar', palette='rocket')
plt.title('Price of Airline based on Departure Time')
plt.xlabel('Departure Time')
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.catplot(x='arrival_time',y='price',data=data, kind='bar', palette='mako')
plt.title('Price of Airline based on Arrival Time')
plt.xlabel('Arrival Time')
plt.show()
```

```python
sns.relplot(x='arrival_time', y='price', data=data, col='departure_time', kind='line')

plt.show()
```
